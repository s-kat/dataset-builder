```python
import json
total = json.load(open('out.json', 'r'))
```

```python
# total files
print(len(total))
```

```python
# total functions
s = 0
funcs = {}

for cur_funcs in total.values():
    for func_name in cur_funcs:
        funcs[func_name] = cur_funcs[func_name]
print(len(funcs))
```

```python
funcs['validate']
```

```python
total_args = []
bad_funcs = []

for func_name in funcs:
    for args in funcs[func_name]['args']:
        if 'self' in args['func_args'].keys():
            bad_funcs.append(func_name)
            break
        for i in args['func_args'].values():
            total_args.append(i)
```

```python
print(len(set(bad_funcs)))
```
